```python
import json
import os
```

```python
with open("/media/bnrc2/_backup/dataset/golf_keypoint_predictions_all.json","r") as f:
    js = json.load(f)
```

```python
import pandas as pd
```

```python
fi = pd.DataFrame(js)
import cv2
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
colors = [
            [ 0, 0,255], [0, 255, 0], [255,0,0], [0, 245, 255], [255, 131, 250], [255, 255, 0],
            [0, 0, 255], [0, 255, 0], [255, 0, 0], [0, 245, 255], [255, 131, 250], [255, 255, 0],
                  [0, 0, 0], [255, 255, 255], [255, 0, 0], [0, 255, 0], [0, 0, 255],[128,128,128]]
joints = ["rShoulder", "rElbow", "rWrist", "lShoulder", "lElbow", "lWrist", "rhip","rknee","rankle",
                       "lhip","lknee","lankle","head","neck","torso"]
line = [[0,1],[1,2],[3,4],[4,5],[6,7],[7,8],[9,10],[10,11],[12,13],[13,14],[6,14],[9,14],[0,13],[3,13]]
limb = ['left low arm','left up arm','right low arm','right low arm',"Left leg","Left thigh","right leg","right thigh"]
limb_num = [[1,2],[0,1],[4,5],[3,4],[7,8],[6,7],[11,10],[10,9]]
offset = [[-100,-50],[-100,+50],[50,0],[50,20],[-50,0],[-50,-20],[50,0],[50,20], ]
```

```python
def distance(a,b):
    return np.sqrt(np.power(a[0] - b[0],2) + np.power(a[1] - b[1],2))   
```

```python
father_path = "/media/bnrc2/_backup/dataset/2D_1710/"
out_path =  "/media/bnrc2/_backup/dataset/2D_1710/citens4/"
font = cv2.FONT_HERSHEY_PLAIN
#font=cv2.InitFont(cv.CV_FONT_HERSHEY_SCRIPT_SIMPLEX, 1, 1, 0, 3, 8)
for index, row in fi.iterrows():   # 获取每行的index、row
    img_path = os.path.join(father_path,row["image_id"][5:])
    img = cv2.imread(img_path)
    
    out_folder = os.path.join(out_path,row["image_id"].split("/")[1])
    file_name = row["image_id"].split("/")[-1]
    
    if not os.path.isdir(out_folder):
        os.mkdir(out_folder)
    #plt.imshow(img)
    anno = row["keypoint_annotations"]
    for key in anno.keys():
        coord = np.reshape(anno[key],[14,3])
        
        hip = np.average(np.stack([coord[6],coord[9]],axis=0),axis=0)
        hip = np.transpose(np.expand_dims(hip,-1))
        coord = np.concatenate([coord,hip],axis=0)
        coord = coord.astype(np.int32)
        all_length = distance(coord[12],coord[13]) +  distance(coord[14],coord[13]) +        max(distance(coord[9],coord[10]) + distance(coord[11],coord[10]), 
            distance(coord[6],coord[7] + distance(coord[7],coord[8]))) 
        shape = img.shape
        for i in range(len(limb)):
            dis = distance(coord[limb_num[i][0]],coord[limb_num[i][1]])
            x,y = (coord[limb_num[i][0]][0] + coord[limb_num[i][1]][0]) / 2,(coord[limb_num[i][0]][1] + coord[limb_num[i][1]][1]) / 2 
            txt =  "%.2f" % ( dis/ all_length * 180)+"cm"
            cv2.putText(img,txt ,(int(x+offset[i][0]),int(y+offset[i][1])), font, 2,(0,255,255),5,cv2.LINE_4)
        
        for i in range(15):
            cv2.circle(img,(int(coord[i][0]),int(coord[i][1])),5, colors[i], -1)
        for j in range(len(line)):
            cv2.line(img,(coord[line[j][0]][0],coord[line[j][0]][1]),(coord[line[j][1]][0],coord[line[j][1]][1]),colors[14],5)
        cv2.imwrite(os.path.join(out_folder,file_name), img)
        break
```
